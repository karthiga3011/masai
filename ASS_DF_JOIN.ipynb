```python
import pandas as pd
import numpy as np
```

```python
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}
```

```python
enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}
```

```python
scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}
```

```python
students = pd.DataFrame(students_data)
enrollments = pd.DataFrame(enrollments_data)
scores = pd.DataFrame(scores_data)
print("Original Students DataFrame\n", students)

print("Original Students DataFrame\n", enrollments)

print("Original Students DataFrame\n", scores)
```

```python
for col in students.columns:
    null_count = students[col].isnull().sum()
    percentage = (null_count / len(students)) * 100
    print(f"Column: {col}, Nulls: {null_count} percentage: {percentage}")
```

```python
students['city'] = students['city'].fillna("Unknown")
students = students.dropna(subset=['name'])

print("\nCleaned Students DataFrame\n", students)
```

```python
inner_join = pd.merge(students, enrollments, on='student_id', how='inner')

print("\nInner Join Result")
print(inner_join)

print("Total Students:", len(inner_join))
```

```python
excluded = set(students.student_id) - set(enrollments.student_id)
print("Excluded students:", list(excluded))
```

```python
left_join = pd.merge(students, enrollments, on='student_id', how='left')

print("\nLeft Join Result")
print(left_join)

print("Total Rows:", len(left_join))
```

```python
null_courses = left_join[left_join['course_name'].isnull()]
print("\nStudents with null course_name")
print(null_courses[['student_id','name']])
```

```python
right_join = pd.merge(students, enrollments, on='student_id', how='right')

print("\nRight Join Result")
print(right_join)

print("Total Rows:", len(right_join))
```

```python
missing_students = right_join[right_join['name'].isnull()]
print("\nStudent IDs without student data")
print(missing_students['student_id'])
```

```python
outer_join = pd.merge(students, enrollments, on='student_id', how='outer')

print("\nOuter Join Result")
print(outer_join)

print("Total Rows:", len(outer_join))
```

```python
outer_join = pd.merge(students, enrollments, on='student_id', how='outer')

print("\nOuter Join Result")
print(outer_join)

print("Total Rows:", len(outer_join))
```

```python
missing_rows = outer_join[(outer_join['name'].isnull()) | (outer_join['course_name'].isnull())]

print("\nRows with Missing Name or Course")
print(missing_rows)
```

```python
outer_indicator = pd.merge(students, enrollments, on='student_id', how='outer', indicator=True)

print("\nMerge Source Distribution")
print(outer_indicator['_merge'].value_counts())
```

```python
score_dict = scores.set_index('student_id')['exam_score'].to_dict()

students['exam_score'] = students['student_id'].map(score_dict)

print("\nStudents with Scores")
print(students[['student_id','name','exam_score']])
```

```python
def auto_merge(df1, df2, join_type, key_column):

    merged_df = pd.merge(df1, df2, how=join_type, on=key_column)

    result = {
        'result_df': merged_df,
        'row_count': len(merged_df),
        'join_type': join_type
    }

    return result
```

```python
test1 = auto_merge(students, enrollments, 'inner', 'student_id')

print("\nJoin Type:", test1['join_type'])
print("Rows:", test1['row_count'])
print(test1['result_df'].head())
```

```python
test2 = auto_merge(students, enrollments, 'left', 'student_id')

print("\nJoin Type:", test2['join_type'])
print("Rows:", test2['row_count'])
print(test2['result_df'].head())
```
